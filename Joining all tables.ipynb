```python
#First merge labs and normalised drugrates
df1 = pd.read_csv("labs_updated.csv")
df2 = pd.read_csv("drugrate_norm.csv")

# creating a empty bucket to save result
df_result = pd.DataFrame(columns=(df1.columns.append(df2.columns)).unique())
df_result.to_csv("merge_labs_drugrate.csv",index_label=False)

# deleting df2 to save memory
del(df2)

def preprocess1(x):
    df2=pd.merge(df1,x, on="patientunitstayid")
    df2.to_csv("merge_labs_drugrate.csv",mode="a",header=False,index=False)

reader = pd.read_csv("drugrate_norm.csv", chunksize=1000) # chunksize depends with you colsize

[preprocess1(r) for r in reader]
```
